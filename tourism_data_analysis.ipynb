```python
import pandas as pd
df = pd.read_csv('tourism/data.csv', usecols=['lat', 'lon', 'type', 'subtype'])
```

## Plot Dataframe

```python
# All locations belonging to 'human activity'
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df['lon'], df['lat'], zorder=1, alpha= 0.2, c='b', s=1)
ax.set_title('Locations in Galicia, Spain')

# All tourism locations
df_tourism = df.loc[df['type'] == 'tourism']
num_tourism_entries = len(df_tourism)
df_tourism = df_tourism.reset_index(drop=True)

fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df_tourism['lon'], df_tourism['lat'], zorder=1, alpha= 0.2, c='b', s=5)
ax.set_title('Tourism locations in Galicia, Spain')
```
